Chaine Rule: P(A, B, C) = P(A | B, C) * P(B | C) * P(C)

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

G.add_nodes_from(["A", "B", "C"])
G.add_edges_from([("C", "B"), ("B", "A"), ("C", "A")])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue")
plt.title("Chain Rule for Probability")
plt.show()
```

```python
G = nx.DiGraph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D')])

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, arrows=True)
plt.show()
```

```python
def calculate_marginal_B(P_A, P_B_given_A, P_C_given_A, P_D_given_B_C):

    P_B = 0

    for a in P_A.keys():

        for c in P_C_given_A[a].keys():

            P_B += P_A[a] * P_B_given_A[a] * P_C_given_A[a][c] * sum(P_D_given_B_C[b][c] for b in P_B_given_A[a].keys())

    return P_B
```
